```python
import matplotlib.pyplot as plt
import csv

with open('valid_progress.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='"')
    epochs = []
    rmse = []
    sre = []
    uqi = []
    for i, row in enumerate(spamreader):
        if i == 0:
            continue
        epochs.append(row[0])
        rmse.append(float(row[1]))
        sre.append(float(row[2]))
        uqi.append(float(row[3]))
```

```python
plt.xlabel("# epoch")
plt.ylabel("RMSE")
plt.plot(epochs,rmse, 'r+', label="RMSE over epoch")
plt.plot([13.767030328673172]*len(rmse), 'b', label="Avg. RMSE for DL model on Testset")
plt.plot([12.840806320121528]*len(rmse), 'g', label="Avg. RMSE for Bilinear Interpolation on Testset")
plt.legend(loc="best")
plt.savefig('RMSE_val.png')
plt.show()
```

```python
plt.ylabel("SRE")
plt.plot(epochs,sre, 'r+', label="SRE over epoch")
plt.plot([18.900307461410126]*len(rmse), 'b', label="Avg. SRE for DL model on Testset")
plt.plot([20.56455829590438]*len(rmse), 'g', label="Avg. SRE for Bilinear Interpolation on Testset")
plt.legend(loc="best")
plt.savefig('SRE_val.png')
plt.show()
```

```python
plt.ylabel("UQI")
plt.plot(epochs,uqi, 'r+', label="UQI over epoch")
plt.plot([0.11981306826719888]*len(uqi), 'b', label="Avg. UQI for DL model on Testset")
plt.plot([0.12816210999705666]*len(uqi), 'g', label="Avg. UQI for Bilinear Interpolation on Testset")
plt.legend(loc="best")
plt.savefig('UQI_val.png')
plt.show()
```

```python
with open('training_progress.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='"')
    epochs = []
    batches = []
    cl = []
    al = []
    pl = []
    acl = []
    aal = []
    apl = []
    for i, row in enumerate(spamreader):
        if i == 0:
            continue
        epochs.append(float(row[0]))
        batches.append(float(row[1]))
        cl.append(float(row[2]))
        al.append(float(row[3]))
        pl.append(float(row[4]))
        acl.append(float(row[5]))
        aal.append(float(row[6]))
        apl.append(float(row[7]))

print(cl)
```

```python
plt.xlabel("# epoch")
plt.ylabel("Content loss")
plt.plot(epochs,cl, 'r+', label="Content loss")
plt.plot(epochs,acl, 'b-', label="Running average content loss")
plt.legend(loc="best")
plt.savefig('cl_train.png')
plt.show()
```

```python
plt.xlabel("# epoch")
plt.ylabel("Adversarial loss")
plt.plot(epochs,al, 'r+', label="Adversarial loss")
plt.plot(epochs,aal, 'b-', label="Running average adversarial loss")
plt.legend(loc="best")
plt.savefig('al_train.png')
plt.show()
```

```python
plt.xlabel("# epoch")
plt.ylabel("Perceptual loss")
plt.plot(epochs,pl, 'r+', label="Perceptual loss")
plt.plot(epochs,apl, 'b-', label="Running average perceptual loss")
plt.legend(loc="best")
plt.savefig('pl_train.png')
plt.show()
```
